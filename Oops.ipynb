## <span style="color:orange"> Object-Oriented Programming (OOP) in Python:<span>

OOP is a programming paradigm that organizes code into objects, which bundle data (attributes) and behaviors (methods) together.

Why Use OOP?

*  Encapsulation  Group related data & methods together.
*  Inheritance  Reuse & extend existing functionality.
*  Polymorphism  Use the same interface for different data types.
*  Code Reusability & Maintainability  Reduces duplication.

```python
class Car:
    def __init__(self ,name, model,year):
        self.name=name
        self.model=model
        self.year=year
    def carinfo(self):
        print(f"car name is {self.name} model is {self.model} and year is {self.year}")

mycar= Car("jaguar","mts",2025)
mycar.carinfo()
```

## Class Variable vs Instance Variable

```python
#Instance Variables (Object-Specific)
class Dog:
    def __init__(self, name):
        self.name=name  # Instance variable (object-specific)

dog1=Dog("Buddy")
dog2=Dog("Max")
print(f"dog1: {dog1.name}")  # Output: Buddy   
print(f"dog2: {dog2.name}")  # Output: Max


#-----Class Variables (Shared Across All Instances)-----------------------------------------
class Cat:
    
    breed="Siamese"  # Class variable (shared across all instances)
    def __init__(self,name):
        self.name=name
        

cat1= Cat("Persian")
cat2= Cat("fusya")
print(f"cat1: {cat1.name} and {cat1.breed}  ")  # Output: Persian
print(f"cat2: {cat2.name} and {cat2.breed}")  # Output: Siamese
```

## <span style="color:pink">Instance Methods,<span style="color:yellow"> Class Methods <span style="color:lightgreen"> Static Methods<span>

 When to Use?

* Instance Methods  Modify instance data.

* Class Methods  Modify class-wide data.

* Static Methods  Utility functions unrelated to class attributes.

```python
# Instance methods( operate on objects)
class car:
    def __init__(self,brand):
        self.brand=brand # Instance variable

    def showbrand(self): # Instance method
        return f"car brand is {self.brand}"
    
mybrand=car("BMW")
print(mybrand.showbrand())

#------------Class methods (Operate on Class-Level Data)-------------------
class Car:
    wheels = 4  # Class variable

    @classmethod
    def change_wheels(cls, new_wheels):
          cls.wheels = new_wheels

Car.change_wheels(6) # Changing class variable using class method
print(Car.wheels)  # 6

#-----------Static methods (utility functions) (independent of class or instance)---------------
class join:
    def t(self):
        return f"string"  # Instance method
    
    @staticmethod
    def concat(str1,str2):  # Static method
        return str1 + str2
    
    def l(str1,str2): # Static method
        return str1.lower() + str2.lower()
    
    
    
    
# Calling static method without creating object,using class name.
strjoin=join.concat("Hello, ","World!")
print(strjoin)  # Output: Hello, World!
print(join.l("HELLO, ","WORLD!"))  # Output: hello, world!
print(join().t()) # here I have created object to call instance method.
```
