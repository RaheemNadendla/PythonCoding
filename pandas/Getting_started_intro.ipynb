## <span style="color:Green">Intro to pandas:<span>
* When working with tabular data, such as data stored in spreadsheets or databases, pandas is the right tool for us.
 pandas will help us to explore, clean, and process our data. In pandas, a data table is called a DataFrame.

<span style="color:red"> What is pandas?<span>

 *---> pandas is a Python package providing fast, flexible, and expressive data structures designed to make working with
 “relational” or “labeled” data both easy and intuitive. It aims to be the fundamental high-level building block for doing
 practical, real-world data analysis in Python. Additionally, it has the broader goal of becoming the most powerful
 and flexible open source data analysis/manipulation tool available in any language.
