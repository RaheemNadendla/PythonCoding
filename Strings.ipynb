```python
s1="hello!" #double  quote string
s2='hello!' #single quote string
s3="""hello!""" #triple double quote string

print(s1[0]) #indexing
print(len(s1)) #length of string
print(s1+' rox!') #concatenation
print(s1[0:3]) #slicing-- it will print index 0,1,2 only.
print(s1[0:5:2]) #slicing with step--it will print index 0,2,4 only.
```

## <span style="color:orange">String methods:<span>https://docs.python.org/3/library/stdtypes.html#string-methods

```python
s="london"
su="LONDON"
par="raw string to ignore escape characters"
add="The sum of 1 + 2 is {} and the sum of 3 + 4 is {}."
print(f"{s.capitalize()}:----> .capitalize() returns first character in capital case and the rest characters will be in lower case.") 
print(f"{s.upper()}:----> .upper() returns all characters in upper case.")
print(f"{s.lower()}:----> .lower() returns all characters in lower case.")
print(f"{su.casefold()}:----> .casefold() returns all characters in lower case, casefold is more aggressive than lower.")
print(f"{par.title()}:----> .title() returns first character of each word in capital case and the rest characters will be in lower case.")
print(f"{s.center(12,"*")}:----> .center() returns the string centered in a field of given width, padding with the specified character.")
print(f"{s.ljust(12,'*')}:----> .ljust() returns the string left justified in a field of given width, padding with the specified character.")
print(f"{s.rjust(12,'*')}:----> .rjust() returns the string right justified in a field of given width, padding with the specified character.")
print(f"{par.count('r',3,19)}:----> .count() returns the number of occurrences of a substring in the string, within the specified range.")
print(f"{s.endswith('on')}:----> .endswith() returns True if the string ends with the specified suffix, otherwise False.")
print(f"{par.endswith('re',14,20)} :----> .endswith() with range returns True if the string ends with the specified suffix within the specified range, otherwise False.")
print(f"{add.format(1+2,3+4)}:---> # .format() method to format strings with positional arguments")
```
